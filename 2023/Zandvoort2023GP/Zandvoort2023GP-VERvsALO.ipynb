```python
import fastf1 as ff1
import fastf1.plotting

from matplotlib import pyplot as plt

# misc_mpl_mods option enables minor grid lines which clutter plots
ff1.plotting.setup_mpl(misc_mpl_mods=False)
```

```python
session = ff1.get_session(2023, 13, 'R')
session.load()
```

```python
ver_lap = session.laps.pick_driver('VER').pick_fastest()
alo_lap = session.laps.pick_driver('ALO').pick_fastest()

ver_tel = ver_lap.get_car_data().add_distance()
alo_tel = alo_lap.get_car_data().add_distance()
```

```python
rbr_color = ff1.plotting.team_color('RBR')
amr_color = ff1.plotting.team_color('AMR')

fi, ax = plt.subplots()
ax.plot(ver_tel['Distance'], ver_tel['Speed'], color=rbr_color, label='VER')
ax.plot(alo_tel['Distance'], alo_tel['Speed'], color=amr_color, label='ALO')

ax.set_xlabel('Distance in m')
ax.set_ylabel('Speed in km/h')

ax.legend()
plt.suptitle(f"Fastest Lap Comparison \n "
             f"{session.event['EventName']} {session.event.year} Race")

plt.show()
```
