```python
import fastf1 as ff1
import fastf1.plotting

from matplotlib import pyplot as plt

# mplib patches for plotting timedeltas
# also loads ff1s color scheme
ff1.plotting .setup_mpl(misc_mpl_mods=False)
```

```python
ff1.plotting.setup_mpl()
schedule = ff1.get_event_schedule(2023)
schedule
```

```python
gp_nl = schedule.get_event_by_name('FORMULA 1 HEINEKEN DUTCH GRAND PRIX 2023')
gp_nl
```

```python
session = ff1.get_session(2023, 13, 'R')
session.load()
laps = session.laps
```

```python
drivers = session.drivers
print(drivers)
```

```python
drivers = [session.get_driver(driver)["Abbreviation"] for driver in drivers]
print(drivers)
```

```python
stints = laps[["Driver", "Stint", "Compound", "LapNumber"]]
stints = stints.groupby(["Driver", "Stint", "Compound"])
stints = stints.count().reset_index()
stints = stints.rename(columns={"LapNumber": "StintLength"})
print(stints)
```

```python
fig, ax = plt.subplots(figsize=(5,10))

for driver in drivers:
    driver_stints = stints.loc[stints["Driver"] == driver]
    
    previous_stint_end = 0
    for idx, row in driver_stints.iterrows():
        # each row contains the compound name and stint length
        # we can use this info to draw horizontal bars
        plt.barh(
            y=driver,
            width=row["StintLength"],
            left=previous_stint_end,
            color=ff1.plotting.COMPOUND_COLORS[row["Compound"]],
            edgecolor="black",
            fill=True
        )
        
        previous_stint_end += row["StintLength"]

plt.title("2023 Dutch GP Compound Strategies")
plt.xlabel("Lap Number")
plt.grid(False)

# invert y axis so drivers that finish higher are closer to the top
ax.invert_yaxis()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)

plt.tight_layout()
plt.show()
```
